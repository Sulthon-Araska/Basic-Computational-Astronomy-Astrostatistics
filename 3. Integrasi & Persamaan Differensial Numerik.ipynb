## **1. Integrasi Numerik**

### 1.1 Aturan Trapezoidal

```python
#@title **Latihan 1.1 (a)**

from math import sin, pi

def f(x):             #Mendefinisikan yang akan diintegrasi
 return x*sin(x)      #f(x) = x.sin(x)

a = 0                 #Batas bawah integrasi
b = pi/2              #Batas atas inetgrasi
n = 6                 #Jumlah pias/segmen
h = (b-a)/n           #Lebar pias
S = 0.5*(f(a)+f(b))   #Suku pertama dari I/h

for i in range(1,n):
  x = a+i*h           #Batas bawah pias ke-(i+1)
  S += f(x)           #Suku ke-(i+1) dari I/h

Integral = h * S      
print('Integral = %f' % Integral)
```

```python
#@title **Latihan 1.1 (b)**

from math import sin, pi

def f(x):
 return x*sin(x)      #Fungsi yang akan diintegrasi

a = 0                 #Batas bawah integrasi
b = pi/2              #Batas atas inetgrasi
n = 1000              #Jumlah pias/segmen
h = (b-a)/n           #Lebar pias
S = 0.5*(f(a)+f(b))   #Suku pertama dari I/h

for i in range(1,n):
  x = a+i*h           #Batas bawah pias ke-(i+1)
  S += f(x)           #Suku ke-(i+1) dari I/h

Integral = h * S      
print('Integral = %f' % Integral)
```

**Galat mutlak** suatu perhitungan numerik dinyatakan oleh:

**E = |1 - (integral numerik)/(Integral analitik)|**

Dengan begitu, galat mutlak dari kedua integrasi numerik adalah:
1.   Untuk n=6, galat absolut sebesar: **E = 0.5731%**

Jika dibandingkan dari galat mutlaknya, **hasil integrasi numerik akan semakin akurat jika jumlah pias semakin banyak**.

### 1.2 Aturan Simpson 1/3

```python
#@title **Latihan 1.2 (a)**

from math import sin, pi

def f(x):
 return x*sin(x)      #Fungsi yang akan diintegrasi

a = 0                 #Batas bawah integrasi
b = pi/2              #Batas atas inetgrasi
n = 6                 #Jumlah pias/segmen
h = (b-a)/n           #Lebar pias
S = f(a) + f(b)       #Suku pertama & terakhir dari I/(h/3)

for i in range(1,n):
  x = a+i*h           #Batas bawah pias ke-(i+1): x_i
  if (i%2 != 0):      #Untuk i bilangan ganjil
    S += 4*f(x)
  else:               #Untuk i bilangan genap
    S += 2*f(x)

Integral = S*(h/3)
print('Integral = %f' % Integral)
```

**Latihan 1.2 (b)**

Apabila dihitung galatnya, **galat mutlak** menggunakan aturan Simpson 1/3 adalah sebesar **0.0079%**. Jika dibandingkan dengan hasil pada *Latihan 1.1(a)*, **galat integrasi menggunakan aturan Simpson 1/3 jauh lebih kecil** atau lebih akurat.

```python
#@title **Latihan 1.3 (c) Trapezoidal**

from scipy.integrate import trapezoid
import numpy as np

def integrand(x):
 return x*sin(x)

a = 0
b = np.pi/2
ya = integrand(a)
yb = integrand(b)

I = trapezoid([ya,yb], [a,b], dx=1.0, axis=- 1)
print('Hasil integral: I =', I)
```

```python
#@title **Latihan 1.3 (d) Simpson**

from scipy.integrate import simpson
import numpy as np

def integrand(x):
 return x*sin(x)

a = 0
b = np.pi/2
ya = integrand(a)
yb = integrand(b)

I = simpson([ya,yb], [a,b], dx=1.0, axis=- 1, even='avg')
print('Hasil integral: I =', I)
```

## **2. Persamaan Diferensial Biasa**

### 2.1 Metode Euler

```python
#@title **Latihan 2.1**

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Fungsi ideal (analitik)
def y(x): return np.exp((x**2)/2)           #Solusi analitik: y = e^(0.5x^2)
x = np.linspace(0,2,1000)                   #Sampel banyak nilai x
plt.plot(x, y(x), label='Analytical')

#Integrasi Numerik Metode Euler
h = [0.5, 0.2]                              #Variasi ukuran langkah
for i in range(len(h)):                     #Iterasi untuk masing-masing ukuran langkah
  h0 = h[i]
  x0 = np.arange(0,2+h0,h0)                 #Rentang nilai x=[0,20] dengan ukuran langkah
  y0 = [1]                                  #List berisi nilai awal y(0)=1
  def Dy(x,y): return x*y                   #Persamaan differensial: dy/dx = xy
  for i in range(len(x0)-1):                #Iterasi untuk setiap langkah
    Y0 = y0[i] + Dy(x0[i],y0[i])*h0         #Perumusan Metode Euler
    y0.append(Y0)                           #Pengisian nilai-nilai y sampel
  plt.plot(x0, y0, label=('h =', h0))       #Line plot untuk masing-masing variasih
  plt.plot(x0, y0, 'o')                     #Scatter plot untuk masing-masing variasih
  
  #Perbandingan hasil numerik VS analitik
  y = [0 for i in range(len(x0))]
  for i in range(len(x0)):
    y[i] = np.exp((x0[i]**2)/2)
  Tabel = pd.DataFrame({'x':x0, 'y (numerik)':y0, 'y (analitik)':y})
  print('Perbandingan nilai fungsi antara metode numerik Euler (h =', h0, ') dengan metode analitik')
  print(Tabel)
  print('')

#Scatter plot
plt.title('Metode Euler')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

### 2.2 Metode Runge-Kutta Orde 2

```python
#@title **Latihan 2.2**

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Fungsi ideal (analitik)
def y(x): return np.exp((x**2)/2)           #Solusi analitik: y = e^(0.5x^2)
x = np.linspace(0,2,1000)                   #Sampel banyak nilai x
plt.plot(x, y(x), label='Analytical')

#Persamaan diferensial
h = 0.5
x0 = np.arange(0,2+h,h)                     #Rentang nilai x=[0,20] dengan ukuran langkah
y0 = [1]                                    #List berisi nilai awal y(0)=1
def Dy(x,y): return x*y                     #Fungsi Persamaan differensial: dy/dx = xy
#Parameter Metode Runge-Kutta orde 2
A = [1/2, 0, 1/3]
B = [1/2, 1, 2/3]
P = [1, 1/2, 3/4]
Q = [1, 1/2, 3/4]
Variant = ['Heun', 'Midpoint', 'Ralston']
#Algoritma untuk masing-masing Metode Heun, Midpoint, & Ralston
for i in range(len(A)):
  y0 = [1]
  a = A[i]
  b = B[i]
  p = P[i]
  q = Q[i]
  #Iterasi untuk setiap langkah
  for j in range(len(x0)-1):
    k1 = Dy(x0[j]      , y0[j])           #Fungsi parameter k1 
    k2 = Dy(x0[j]+p*h , y0[j]+q*k1*h)     #Fungsi parameter k2
    Y0 = y0[j] + (a*k1 + b*k2)*h          #Perumusan Metode Runge-Kutta orde 2
    y0.append(Y0)                         #Pengisian nilai-nilai y sampel
  plt.plot(x0, y0, 'o')                   #Scatter plot untuk masing-masing variasi metode
  plt.plot(x0, y0, '-', label=Variant[i]) #Line plot untuk masing-masing variasi metode
  #Perbandingan hasil numerik VS analitik
  y = [0 for k in range(len(x0))]
  for k in range(len(x0)):
    y[k] = np.exp((x0[k]**2)/2)
  Tabel = pd.DataFrame({'x':x0, 'y (numerik)':y0, 'y (analitik)':y})
  print('Perbandingan nilai fungsi antara metode numerik', Variant[i], 'orde-2 dengan metode analitik')
  print(Tabel)
  print('')

#Scatter plot
plt.title('Metode Runge-Kutta orde 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

### 2.3 Metode Runge-Kutta Orde 4

```python
#@title **Latihan 2.3**

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Fungsi ideal (analitik)
def y(x): return np.exp((x**2)/2)           #Solusi analitik: y = e^(0.5x^2)
x = np.linspace(0,2,1000)                   #Sampel banyak nilai x
plt.plot(x, y(x), label='Analytic')

#Persamaan diferensial dengan metode numerik
h = 0.5
x0 = np.arange(0,2+h,h)                   #Rentang nilai x=[0,20] dengan ukuran langkah
y0 = [1]                                  #List berisi nilai awal y(0)=1
def Dy(x,y): return x*y                   #Fungsi Persamaan differensial: dy/dx = xy
#Parameter Metode Runge-Kutta orde 4
for i in range(len(x0)-1):                #Iterasi untuk setiap langkah
  K1 = h*Dy(x0[i] , y0[i])                #Fungsi Parameter K1
  K2 = h*Dy(x0[i] + h/2 , y0[i]  + K1/2)  #Fungsi parameter K2
  K3 = h*Dy(x0[i] + h/2 , y0[i]  + K2/2)  #Fungsi parameter K3
  K4 = h*Dy(x0[i] + h , y0[i]  + K3)      #Fungsi parameter K4
  Y0 = y0[i] + (K1 + 2*K2 + 2*K3 + K4)/6  #Perumusan Metode Runge-Kutta orde 4
  y0.append(Y0)                           #Pengisian nilai-nilai y sampel
plt.plot(x0, y0, 'o')                     #Scatter plot untuk masing-masing variasi metode
plt.plot(x0, y0, '-', label='Numeric')    #Line plot untuk masing-masing variasi metode

#Perbandingan hasil numerik VS analitik
y = [0 for j in range(len(x0))]
for j in range(len(x0)):
  y[j] = np.exp((x0[j]**2)/2)
Tabel = pd.DataFrame({'x':x0, 'y (numerik)':y0, 'y (analitik)':y})
print('Perbandingan nilai fungsi antara metode numerik Runge-Kutta orde-4 dengan metode analitik')
print(Tabel)

#Scatter plot
plt.title('Metode Runge-Kutta orde 4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

### 2.4 Menggunakan scipy.integrate odeint.

```python
#@title **Latihan 2.4**

from scipy.integrate import odeint  #Mengimpor fitur dari Scipy untuk mengintegralkan persamaan diferensial biasa
from numpy import linspace          #Mengimpor fitur 'linspace' dari library Numpy

dy = lambda y, x: x*y #Persamaan differensial (dy/dx)
y0 = 1                #Nilai awal y(0)=1
x = linspace(0, 2, 5) #Domain nilai x
y = odeint(dy, y0, x) #Integrasi numerik dari persamaan differensial

print(y)  #Menampilkan nilai-nilai sampel y hasil integrasi
```

Dibandingkan dengan metode Euler (Latihan 2.1) & Runge-Kutta (Latihan 2.2 dan 2.3), **fitur *scipy.interpolate.odeint* memberikan hasil fungsi (y) yang lebih akurat**. Hal ini terlihat dari hasilnya lebih mendekati nilai benar (hasil analitik) untuk nilai-nilai x yang diberikan, bahkan dengan presisi yang lebih tinggi (angka desimal lebih banyak).
