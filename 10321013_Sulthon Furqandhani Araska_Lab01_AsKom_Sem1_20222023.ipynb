Suatu persamaan fungsi rotasi kritis bintang memenuhi bentuk persamaan fungsi

$$f(x) = x - \frac{\omega^2r_p^3}{2GM}\;x^3 - 1$$

dengan $x = r/r_p$, dan $r$ adalah jejari bintang arah ekuator, sedangkan $r_p$ adalah jejari bintang arah kutub (indeks `p` untuk *pole*). Konstanta $G$ sebagaimana diketahui bersama adalah sebesar $6,\!673 \times 10^{-11} \mbox{ m}^3\mbox{ kg}^{-1}\mbox{ s}^{-2}$ (dalam *coding* dapat ditulis: `6.673E-11`). Untuk modul ini, massa bintang adalah serupa Matahari, $M = M_{\odot} = 1,\!989 \times 10^{30} \mbox{ kg}$ (dalam *coding* dapat ditulis: `1.989E30`), jejari arah kutub $r_p = 6,\!95 \times 10^8\mbox{ m} \neq R_{\odot} $ (dalam *coding* dapat ditulis: `6.95E8`), dan kecepatan sudut $\omega = 2,\!424068 \times 10^{-4}\mbox{ rad s}^{-1}$ (dalam *coding* dapat ditulis: `2.424068E-4`) yang setara dengan periode rotasi $0,\!3$ hari.

Akar persamaan fungsi rotasi kritis bintang tersebut akan dipecahkan, yaitu untuk:

$$f(x) = x - \frac{\omega^2r_p^3}{2GM}\;x^3 - 1 = 0$$

Jika dimisalkan solusinya adalah $x =x_c = r_c/r_p$, hal ini menyatakan situasi perbandingan jejari arah ekuator sebesar $r_c$ terhadap jejari arah kutub, $r_p$, di saat kritis menjelang pecahnya kesetimbangan gaya-gaya yang bekerja.

### Definisi fungsi

```python
def f(x):
    G, M, rp, w = 6.673E-11, 1.989E30, 6.95E8, 2.424068E-4
    return x - ((w**2)*(rp**3)*(x**3))/(2*G*M) - 1
```

### Memuat modul-modul python yang diperlukan

```python
import scipy.optimize as sco
import numpy as np
import matplotlib.pyplot as plt
```

### Menyiapkan domain $x$ dan $y$ untuk fungsi menghitung dan plot $f(x)$

Sebagai contoh:

```python
xleft2plot, xright2plot, pointsNumber = 10, 10, 1
x = np.linspace(xleft2plot, xright2plot, pointsNumber)
y = f(x)
```

### Plot esensial untuk $f(x)$

```python
plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')
plt.grid()

plt.plot(x,y)
```

## TUGAS 1

Lakukan dan tunjukkan contoh eksperimen numerik yang Anda lakukan di bawah ini dengan menggunakan metode bisection:

```python
#Definisi Fungsi
def f(x):
    G, M, rp, w = 6.673E-11, 1.989E30, 6.95E8, 2.424068E-4
    return x - ((w**2)*(rp**3)*(x**3))/(2*G*M) - 1

#Memuat modul-modul
import scipy.optimize as sco
import numpy as np
import matplotlib.pyplot as plt

#Menyiapkan domain untuk menghitung fungsi
xleft2plot, xright2plot, pointsNumber = -5, 5, 1000
x = np.linspace(xleft2plot, xright2plot, pointsNumber)
y = f(x)

#Plot esensial untuk f(x)
plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')
plt.grid()
plt.plot(x,y)

#Mengaproksimasi nilai akar menggunakan metode bisection
x = sco.bisect(f, -4, -4.1, maxiter=70, full_output=True, disp=True)
x
```

```python
#Dengan nilai jejari kutub (rp) yang diberikan,
rp = 6.95E8
x = sco.bisect(f, -4, -4.1, maxiter=70, full_output=False, disp=True)

#...diperoleh nilai jejari ekuatorial:
rc = x*rp
rc
```

## TUGAS 2

Lakukan dan tunjukkan contoh eksperimen numerik yang Anda lakukan di bawah ini dengan menggunakan metode Newton-Raphson:

```python
#Definisi Fungsi
def f(x):
    G, M, rp, w = 6.673E-11, 1.989E30, 6.95E8, 2.424068E-4
    return x - ((w**2)*(rp**3)*(x**3))/(2*G*M) - 1

#Memuat modul-modul
import scipy.optimize as sco
import numpy as np
import matplotlib.pyplot as plt

#Menyiapkan domain untuk menghitung fungsi
xleft2plot, xright2plot, pointsNumber = -5, 5, 1000
x = np.linspace(xleft2plot, xright2plot, pointsNumber)
y = f(x)

#Plot esensial untuk f(x)
plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')
plt.grid()
plt.plot(x,y)

#Mengaproksimasi nilai akar menggunakan metode Newton-Raphson
x = sco.newton(f, 0, fprime=None, args=(), tol=1e-07, maxiter=70, rtol=0.0, full_output=True, disp=True)
x
```

```python
#Dengan nilai jejari kutub (rp) yang diberikan,
rp = 6.95E8
x = sco.newton(f, 0, fprime=None, args=(), tol=1e-07, maxiter=70, rtol=0.0, full_output=False, disp=True)

#...diperoleh nilai jejari ekuatorial:
rc = x*rp
rc
```
