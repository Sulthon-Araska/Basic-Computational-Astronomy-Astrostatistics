```python
import pandas as pd  
import scipy.interpolate
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
# Data dari Katalog

Katalog = pd.read_excel('Katalog cepheids.xlsx')
```

```python
# IR Cepheid

Data = pd.read_excel('IR Cep.xlsx')

Flux = Data['SAP_FLUX']
time = Data['time']

x = np.linspace(time[0], time[len(Data) - 1], 500)

y = scipy.interpolate.interp1d(time, Flux, kind = 'cubic')

print("Plot Flux vs time")
plt.plot(x, y(x)) 
plt.xlabel('time (JD)')
plt.ylabel('Flux')
plt.show()

print("Nilai Rata-rata Flux IR Cepheid = " + str(Flux.mean()))
print("Nilai Standar Deviasi Flux IR Cepheid = " + str(Flux.std()))
print('')
```

```python
Data = pd.read_excel('Flux IR.xlsx')
Fase = Data['Fasa']
Flux = Data['Flux Rata2']
SD = Data['Standar Deviasi']
fig, ax = plt.subplots()

ax.errorbar(Fase, Flux, yerr=SD, fmt='-o')
plt.xticks(np.arange(0, 1 , 0.1))
 
ax.set_xlabel('Fase')
ax.set_ylabel('Flux Rata-rata')
ax.set_title('Diagram Fase')
plt.show()
```
